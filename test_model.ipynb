```python
import numpy as np
import tensorflow as tf
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
import seaborn as sns


def showMatriz(matriz):
    print(matriz.shape)
    sns.heatmap(matriz, cmap="YlGnBu")
    plt.show()
```

```python
# charge model
model = tf.keras.models.load_model('model.h5')
```

```python
img = tf.keras.preprocessing.image.load_img(
    'test2.png', target_size=(224, 224)
)

# show image with plt
plt.imshow(img)
```

```python
    # take portion of image
img = cv.imread('test2.png')
img = img[130:260, 0:188]

print(img.shape)

#img =img[20:-20, 60:-60, ]
# resize image to 224x224
img = cv.resize(img, (28, 28))

# shape (28,28,3) is 28 x 28 array and 3 for 3 channels (RGB) (3d array)

# transform to gray scale (2d array)
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img = cv.bitwise_not(img)

# show image with plt
showMatriz(img)
```

```python
# reshape image to 1 dimension
input_data = img.reshape(1, 784)

# to float32
input_data = tf.cast(input_data, tf.float32)
showMatriz(input_data)

# transform shape (150, 672) to (1, 784)
#input_data = input_data.reshape(-1, 784)

#print(input_data.shape)
```

```python
res = model(input_data)

# print result
print(tf.cast(res.numpy(), tf.int32))
showMatriz(res)

# ndim is dimension of array
res.ndim
```
